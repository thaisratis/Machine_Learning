```python
import matplotlib.pyplot as plt
import numpy as np
```

# Line plot

```python
x = np.array([1,2,5])
y = np.array([2,3,7])
```

```python
plot_title = "Line plot"
x_axis = "X axis"
y_axis = "Y axis"
```

```python
# Put the legends
plt.title(plot_title)
plt.xlabel(x_axis)
plt.ylabel(y_axis)

# Perform the plot
plt.plot(x,y)
plt.show()
```

# Bar plot

## A simple barplot with single bars

```python
x = np.array([1,2,3,4,5])
y = np.array([2,3,7,1,0]) #axis lenght
```

```python
plot_title = "Simple bar plot"
x_axis = "Number"
y_axis = "Lenght"
```

```python
# Put the legends
plt.title(plot_title)
plt.xlabel(x_axis)
plt.ylabel(y_axis)

# Perform the plot
plt.bar(x,y)
plt.show()
```

## Bar plot with multiple bars to comparisons

```python
x1 = np.array([1,3,5,7,9])
y1 = np.array([2,3,7,1,0]) #axis lenght


x2 = np.array([2,4,6,8,10])
y2 = np.array([5,1,3,7,4]) #axis lenght
```

```python
#bar plot with legends
plt.bar(x1,y1, label = "Grup 1") #label = Lengend's title
plt.bar(x2,y2, label = "Grup 2")
plt.legend()
plt.title("Bar plot")
plt.xlabel("Number")
plt.ylabel("Bar lenght")
plt.show()
```

# Scatterplot

```python
x = np.array([1,2,3,4,5])
y = np.array([2,3,7,1,0]) # Axis lenght
```

```python
plot_title = "Scatterplot"
x_axis = "Number"
y_axis = "Lenght"
```

## Simple scatterplot

```python
#legends
plt.title(plot_title)
plt.xlabel(x_axis)
plt.ylabel(y_axis)

plt.scatter(x,y)
plt.show()
```

## Putting a line to link the points

```python
#legends
plt.title(plot_title)
plt.xlabel(x_axis)
plt.ylabel(y_axis)

plt.plot(x,y, color = "#990000", linestyle = "--") #Put a line to link the points
plt.show()
```

## Create each circle with a specific lenght

```python
x = np.array([1,2,3,4,5])
y = np.array([2,3,7,1,0]) # Axis lenght
z = np.array([200,25,400,3300,100]) # Circle's lenght
```

```python
#color = Point's color (black), marker = Marker type (.), s = marker's lenght (circle's lenght)
plt.scatter(x,y, label = "My points", color = "k", marker = ".", s = z)

plt.plot(x,y, color = "#990000", linestyle = "--") #Put a line to link the points
plt.legend()
plt.show()
```

# Boxplot

```python
#Build an aleatory numpy array with: number 1 is the minumum value, number 5 is tha maximum and the vector contains 10 numbers
aleatory_vector = np.random.randint(low=1, high=5, size=10)

plt.boxplot(aleatory_vector)
plt.title("Boxplot")
plt.show()
```

# Using datasets

```python
import pandas as pd
df = pd.read_csv("datasets/populacao_brasileira.csv", sep = ";")
```

```python
df.head()
```

```python
df['year'] = df['year'].astype(int)
df['population'] = df['population'].astype(int)
```

```python
plt.bar(df['year'],df['population'], color = "#e4e4e4") # bar plot
plt.plot(df['year'],df['population'], color = "k", linestyle = "--") # line plot

plt.title("Growth of the Brazilian population (1980-2016)")
plt.xlabel("Year")
plt.ylabel("Population")
plt.show()
```

```python
#boxplot
plt.boxplot(df['population'])
plt.xlabel("Population")
plt.ylabel("Numbers")
plt.title("Boxplot")
plt.show()
```
